#### Importing table from URL to a dataframe

```python
import pandas as pd

URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
d = pd.read_html(URL)
df = d[0]
df
```

#### Removing rows from column Borough where value is "Not Assigned"

```python
data = df[df.Borough != "Not assigned" ]
data
```

```python
data.shape
```

#### Reading CSV file for Geospatial Coordindates

```python
postalcodedata = pd.read_csv('C://Users//yadasunn//Downloads//Geospatial_Coordinates.csv')
postalcodedata.head()
```

#### Merging the first dataframe with geospatial coordinates dataframe

```python
finaldata = data.merge(postalcodedata , on = 'Postal Code' , how = 'inner')
finaldata.head()
```

```python
finaldata.shape
```

```python
data_toronto = finaldata[finaldata['Borough'].str.contains('Toronto',regex = False)]
data_toronto
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
!pip install scikit-learn==0.23.2
from sklearn.cluster import KMeans
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors
```

#### Visulaizing all the Toronto Neighbourhood using Folium

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(data_toronto['Latitude'],data_toronto['Longitude'],data_toronto['Borough'],data_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

#### Using KMeans clustering for clustering the neighbourhood of Toronto

```python
k=5
toronto_clustering = data_toronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
data_toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
data_toronto
```

#### Using the above data frame to visualize the Neighbourhood of Toronto clustered

```python
import numpy as np
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)


x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]


markers_colors = []
for lat, lon, neighbourhood, cluster in zip(data_toronto['Latitude'], data_toronto['Longitude'], data_toronto['Neighbourhood'], data_toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
